# Example

Define some routines for the notebook.  Create an array.

```python
x = np.linspace(0, 1)
```

Now plot a line with variable slope.

```python
mpl.rcParams.update(mpl.rcParamsDefault)
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams.update({'xtick.direction': 'in', 'ytick.direction': 'in'})
font = fm.FontProperties(size=20)
fs0=22
fs1=16
def plot_with_slope(m):
    plt.plot(x, m * x**2, label="m = {:.1f}".format(m))
    plt.xlabel("x", fontsize=fs0)
    plt.ylabel(f"y = $m * x^2$", fontsize=fs0)
    plt.legend(fontsize=fs1)
    plt.xticks(fontsize=fs1)
    plt.yticks(fontsize=fs1)
    

#Defining graphic style

    
interact(plot_with_slope, m = widgets.BoundedFloatText(value=1, min=0, max = 10, step=0.1, 
                                                       descriptions='Slope:', disabled=False))
plt.show()
```
